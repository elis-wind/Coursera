```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
import numpy as np

df = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')

#get rid of leap days
df = df[~df['Date'].str.contains("02-29")]

#add 'month_day' column
df['month_day'] = df['Date'].replace("[0-9]{4}-", "", regex = True)

#get the data for the years between 2005 and 2014
df['Date'] = pd.to_datetime(df['Date'])
new = (df['Date'].dt.year >= 2005) & (df['Date'].dt.year <= 2015)
df = df[new]

#get the common temperature
df['Data_Value'] = df['Data_Value'] * 0.1

#get lowest and highest temperatures by day of the year between 2004 and 2014
lowest  = df[(df['Element'] == 'TMIN') & (df['Date'].dt.year != 2015)].groupby('month_day')['Data_Value'].min()
highest = df[(df['Element'] == 'TMAX') & (df['Date'].dt.year != 2015)].groupby('month_day')['Data_Value'].max()

#get lowest and highest temperatures by day for 2015
lowest2015  = df[(df['Element'] == 'TMIN') & (df['Date'].dt.year == 2015)].groupby('month_day')['Data_Value'].min()
highest2015 = df[(df['Element'] == 'TMAX') & (df['Date'].dt.year == 2015)].groupby('month_day')['Data_Value'].max()

#get record break for 2015
highest2015_break = pd.Series(highest2015[highest2015 > highest])
lowest2015_break = pd.Series(lowest2015[lowest2015 < lowest])

#plotting a figue
%matplotlib inline
plt.figure(figsize=(15,12))

plt.plot(highest.values, color='coral', label='2005-2014 Highest Temperatures')
plt.plot(lowest.values, color='deepskyblue', label='2005-2014 Lowest Temperatures')

highest2015_break.reindex(highest.index).plot(marker='o', linestyle='none', color='r', label= '2015 High Record Break')
lowest2015_break.reindex(lowest.index).plot(marker='o', linestyle='none', color='b', label= '2015 Low Record Break')

plt.xticks(range(0, len(highest), 30), highest.index[range(0, len(highest), 30)])
plt.xlabel('Period, month and day', fontsize=15)
plt.ylabel('Temperature, $\degree$ Celsius', fontsize=15)

plt.legend(loc=8, frameon=False, title='', fontsize=12)

plt.title('Record High and Low Temperatures for 2005-2015 in Ann Arbor, Michigan', fontsize=20)

plt.gca().fill_between(range(len(highest)), 
                       highest, lowest, 
                       facecolor='gray', 
                       alpha=0.15)
```
